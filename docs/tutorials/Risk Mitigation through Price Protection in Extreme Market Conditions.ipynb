# 極端な市場状況における価格保護によるリスク軽減

高頻度取引業者やマーケットメーカーにとって、レイテンシーは収益性を維持する上で重要な役割を果たします。
しかし、特に暗号通貨市場では、価格の大幅な変動や市場更新の遅延が一般的です。
これらの不利な条件から見積もりやポジションを保護するためには、Binance が提供するような価格保護メカニズムを採用することが重要です。

https://www.binance.com/en/support/faq/how-to-enable-the-price-protection-function-2b9dc811ce7340469357867122b975dc

> 価格保護は、Binance Futures が提供するもう一つの機能で、極端な市場変動からトレーダーを保護します。この機能は、市場の効率性を悪用し、価格操作を引き起こす悪意のある行為者からトレーダーを保護します。
> 
> 価格保護機能は、最後の価格とマーク価格の間に大きな差があるような異常な市場状況に対して有効です。通常、マーク価格は最後の価格から数セントしか離れていません。しかし、極端な市場状況では、最後の価格がマーク価格から大幅に乖離することがあります。

Binance が強調しているように、先物価格とその基礎となる現物価格の間に大きな差がある場合、それは極端な市場状況を示している可能性があります。これを軽減するためには、先物とその基礎となる現物の最低入札価格および最高売却価格を設定するなど、保守的な価格設定戦略を採用することが重要です。
さらに、先物と基礎となる現物価格の間の価格差の異常を検出することで、ポジションを解消し、通常の市場状況に戻るのを待つことができます。

また、レイテンシー、特にフィードレイテンシーとオーダーレイテンシーを慎重に監視する必要があります。これにより、市場価格の追跡が妨げられ、注文のタイムリーな調整が困難になります。極端な市場状況では、レイテンシーのスパイクが発生することが多く、価格保護を妨げる可能性があるため、そのような状況では市場から撤退することが推奨されます。
